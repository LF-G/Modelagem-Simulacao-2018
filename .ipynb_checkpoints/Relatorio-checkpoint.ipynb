### $$Introdução$$


Para este trabalho, foram realizados experimentos de simulação do Movimento Retilíneo Uniforme (MRU) e do Movimento Retilíneo Uniformemente Variado (MRUV). Através da utilização de cronômetros e de um acelerômetro, foram adquiridos dados para que fosse possível modelar matematicamente os movimentos realizados. Essa modelagem foi, então, comparada com modelos teóricos de referência, a fim de se aferir o quão próxima da realidade ela estava.


### $$Metodologia$$


Os experimentos se baseiam na travessia de 30 metros cronometrados em certas posições e com a utilização de um acelerômetro. Para reproduzir o MRU, decidimos marcar linhas a cada 60 centímetros, onde seriam os passos da caminhada, e controlamos o ritmo através de um metrônomo. Fizemos 3 travessias por pessoa, duas com dois cronômetros nas marcas 10m, 20m e 30m, e uma travessia registrando os tempos de forma alternada nas marcas 5m, 10m, 15m, 20m, 25m e 30m. <br>
Quanto ao MRUV, decidimos que a maneira mais prática seria a de diminuir o intervalo do metrônomo de forma constante, realizando assim uma aceleração nos passos. <br>
A realização do experimento pode ser vista no vídeo: https://youtu.be/9vbi8ogdFyY <br>
Após o experimento, os dados foram analisados e colocados em gráficos juntamente com a formula analítica para facilitar a comparação dos movimentos.

### $$Algoritmo$$



```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

nomes = ['Eugenio', 'Rodrigo', 'Mateus']

#função que grafica os dados do acelerômetro. É necessário que a pasta com os arquivos csv esteja junto do código
def my_plot(arquivo, titulo, nome_eixo_x, nome_eixo_y):
    dados = np.genfromtxt(arquivo, delimiter=';', skip_header=1, names=['t', 'ax', 'ay','az', 'aT'])
    plt.title(titulo)
    plt.xlabel(nome_eixo_x)
    plt.ylabel(nome_eixo_y)
    plt.plot(dados['t'], dados['ay'])
    plt.show()
    
#função que grafica o MRU
def space_time(arquivo, titulo):
    dados=np.genfromtxt(arquivo, delimiter=';',dtype= None, skip_header=1, names=['tempo1', 's','mov'])
    plt.title(titulo)
    plt.xlabel('Tempo(s)')
    plt.ylabel('Espaço(m)')

    #tempo final
    tempof = dados[-1]['tempo1']
    
    #velocidade média
    vel=velMed(tempof)
    print(vel)
    
    # fórmula analítica do MRU, considerando que espaço inicial e velocidade inicial são iguais a 0
    a,=plt.plot(dados['tempo1'],(dados['tempo1'] *vel), label='Analítico' )
    b,=plt.plot(dados['tempo1'], dados['s'],linestyle='',marker='o', label='Experimento')
    
    patch = mpatches.Patch(color='white' ,label='Velocidade Média = ' + format(vel, '.4f'))
    plt.legend(handles=[a,b,patch])
    plt.show()

#função que calcula a velocidade média, sendo que o espaço final é 30
def velMed(tempof):
    return (30/tempof)

#função que calcula a aceleração. A velocidade inicial foi calculada considerando o metrônomo em 60bpm e cada passo sendo 0.6m
def ace(tempof,s):
    return((2*(s-0.6*tempof))/(tempof**2))
  
#função que grafica o MRUV
def vel_time(arquivo, titulo):    
    dados=np.genfromtxt(arquivo, delimiter=';',dtype= None, skip_header=1, names=['tempo1', 's','mov'])

    tempof = dados[-1]['tempo1']
    a= ace(tempof,30)
    t=0
    tem=[]
    spa=[]
    vel=[]
    # o metrônomo estava inicialmente a 60 batidas por minuto (1 batida por segundo).
    # Como cada batida corresponde a um deslocamento de 60 cm, que foi a distância usada
    # entre as marcações no chão, temos uma velocidade inicial de 0.6 m/s

    while t<tempof:
        tem.append(t)
        spa.append(0.6*t + (1/2)*a*t*t)
        vel.append(0.6 + a*t)
        t+=1
        

    vel.append(0.6 + a*tempof)
    tem.append(tempof)
    spa.append(30)
    amed= (vel[-1] - vel[0]) /tempof
    
    fig, y1 =plt.subplots()
    d, =y1.plot(tem,spa, label='Analítico' )
    c, =y1.plot(dados['tempo1'], dados['s'],linestyle='',marker='o', label='Experimento')
    y1.set_xlabel('Tempo(s)')
    y1.set_ylabel('Espaço(m)')
    y1.tick_params('y')
    
    
    y2=y1.twinx()
    y2.set_ylim(0,(1 + vel[-1]))
    b, =y2.plot(tem,vel, label='Velocidade' , color='C2' )
    y2.set_ylabel('Velocidade')
    y2.tick_params('y')
    fig.tight_layout()
   
    patch = mpatches.Patch(color='white' ,label='Aceleração média = ' +  format(amed, '.4f'))
    y1.legend(handles=[d,b,c,patch])
    plt.title(titulo)
    plt.show()
```

### $$ Verificação~do~programa$$ 

```python
# os arquivos .csv usados nesta seção foram fabricados com solução conhecida, assumindo sempre um percurso de 30 m, e
# para o MU: v = 2 m/s, tempo total = 15s
# para o MUV: a = 0,48 m/s, v_0 = 0.6 m/s, tempo total = 10s

space_time('temposres/mu_verif.csv', 'verificação do MU')
vel_time('temposres/muv_verif.csv', 'verificação do MUV')
```

### $$Dados$$

```python
#Às vezes é necessário rodar o programa duas vezes para aparecer os gráficos

for nome in nomes:
     space_time('temposres/' +'tempo'+ nome.lower() + '.csv', 'MU '+ nome )      

for nome in nomes:
     vel_time('temposres/' +'tempo'+ nome.lower() + 'muv'+ '.csv', 'MUV '+ nome )

for nome in nomes:
    for i in range(1,4):
        my_plot('csv/' + nome.lower() + str(i) + '.csv', 'Movimento Uniforme ' + str(i) + ': ' + nome, 'tempo', 'aceleração (eixo z)')

for nome in nomes:
    for i in range(1, 4):
        my_plot('csv/' + nome.lower() + 'muv' + str(i) + '.csv', 'MUV' + str(i) + ': ' + nome, 'tempo', 'aceleração (eixo y)')
```

### $$Análise~e~interpretação$$

Através dos dados, é possível perceber que a metodologia utilizada reproduz um movimento retilíneo uniforme, visto que o espaço aumenta de forma constante. Além disso, nota-se que a formula analítica é uma boa simulação da realidade. A metodologia utilizada para o MRUV também reproduz o movimento e, apesar da aceleração baixa, é possível ver a variação da velocidade representada pela linha verde nos gráficos. Assim, a formula analítica para o movimento simula bem a realidade. Os dados do acelerômetro, apesar dos ruídos causados pelo movimento das mãos, demonstram também que a aceleração foi próxima de 0 nos MRU e proxíma de 0.5 nos MRUV, o que se aproxima da aceleração média calculada de forma analítica.

### $$Crítica$$

A realização do experimento e criação deste relatório fizeram com que o grupo pudesse enteder mais de python e das bibliotecas matplotlib e numpy. Através da utilização do mesmo acelerômetro para todos os experimetos, o grupo evitou um problema, mencionado em sala de aula: o da incompatibilidade de dados quando gravados em diferentes aparelhos. Porém, na hora da simulação do MRUV, notamos que, ao diminuir o tempo entre batidas do metrônomo a cada passo (que era a ideia original do experimento), não tinhamos uma aceleração constante, e sim uma aceleração com variação linear. Para resolver o problema, decidimos alterar o metrônomo a cada segundo. Com isso, notamos outro problema: a dificuldade de dar os passos e de alterar o metrônomo ao mesmo tempo. Resolvemos isso delegando a outro integrante a tarefa de incrementar o metrônomo de forma constante com o uso de um fone sem fio. Após a análise, notamos que isso pode ter gerado um dado inseguro, já que difere muito da média. Outro problema encontrado é o de que, no inicio, assumimos a velocidade inicial como 0 e isso resultava em grandes discrepâncias entre os dados experimentais e a fórmula analítica. Desconsideramos também a interferência no acelerômetro causada pelo movimento das mãos, como visto por exemplo na diferença dos gráficos 1 e 2 do MU realizado pelo Eugenio.
